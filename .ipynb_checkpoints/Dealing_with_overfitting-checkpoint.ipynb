```python
X = df.drop('benign_0__mal_1', axis = 1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))

#Activation Function is sigmoid because this is a classification task
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test, y_test))
```

```python
model_history = pd.DataFrame(model.history.history)
```

```python
model_history
```

```python
model_history.drop(['accuracy', 'val_accuracy'], axis = 1).plot()
```

```python
model_history.drop(['loss', 'val_loss'], axis = 1).plot()
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))

#Activation Function is sigmoid because this is a classification task
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test, y_test), callbacks = [early_stop])
```

```python
model_history = pd.DataFrame(model.history.history)
```

```python
model_history
```

```python
model_history.drop(['loss', 'val_loss'], axis = 1).plot()
```

```python
model_history.drop(['accuracy', 'val_accuracy'], axis = 1).plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dropout(rate = 0.5)) # 0.5 rate means half of the neurons will be randomly turned off during each batch

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(rate = 0.5))

#Activation Function is sigmoid because this is a classification task
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test, y_test), 
          callbacks = [early_stop])
```

```python
model_history = pd.DataFrame(model.history.history)
```

```python
model_history
```

```python
model_history.drop(['loss', 'val_loss'], axis = 1).plot()
```

```python
model_history.drop(['accuracy', 'val_accuracy'], axis = 1).plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
